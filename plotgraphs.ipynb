```python
import matplotlib.pyplot as plt

x=[]
y=[]

nox=int(input("how many values u want to enter for x "))

for i in range(0,nox):
 nx=input("enter the values of x ")
 x.append(nx)

noy=int(input("how many values u want to enter for y "))

for i in range(0,noy):
  ny=input("enter the values of y ")
  y.append(ny)

print("which graph do u want to see")
print("1.line graph")
print("2.bar graph")
print("3.scatter graph")

graph=int(input("enter the choice "))

if graph==1:
 plt.plot(x,y)
 plt.show()

elif graph==2:
 plt.bar(x,y)
 plt.show()

elif graph==3:
  plt.scatter(x,y)
  plt.show()

else:
   print("choice not available")
```
